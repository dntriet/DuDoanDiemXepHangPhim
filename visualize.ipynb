```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu từ file CSV đã được kết hợp
#file_path = 'merged_data.tsv'  # Thay 'path_to_your_file' bằng đường dẫn tới tệp của bạn
file_path = 'D:\Tai_lieu_SV\Hk6\IS252-Khai_thac_du_lieu\DoAn\Code_DoAn/merged_df.csv'
data = pd.read_csv(file_path)
```

```python
# Thống kê mô tả tổng quan về dữ liệu
desc_stats = data.describe(include='all')

# Giá trị cao/thấp nhất
high_low_values = {
    'startYear': {'min': data['startYear'].min(), 'max': data['startYear'].max()},
    'runtimeMinutes': {'min': data['runtimeMinutes'].min(), 'max': data['runtimeMinutes'].max()},
    'averageRating': {'min': data['averageRating'].min(), 'max': data['averageRating'].max()},
    'numVotes': {'min': data['numVotes'].min(), 'max': data['numVotes'].max()},
}

# Giá trị phổ biến/hiếm nhất
common_rare_values = {
    'genres': {'most_common': data['genres'].mode()[0], 'least_common': data['genres'].value_counts().idxmin()},
    'titleLength': {'most_common': data['titleLength'].mode()[0], 'least_common': data['titleLength'].value_counts().idxmin()},
    'numGenres': {'most_common': data['numGenres'].mode()[0], 'least_common': data['numGenres'].value_counts().idxmin()},
}

# Giá trị trung bình và trung vị
average_median_values = {
    'averageRating': {'mean': data['averageRating'].mean(), 'median': data['averageRating'].median()},
    'runtimeMinutes': {'mean': data['runtimeMinutes'].mean(), 'median': data['runtimeMinutes'].median()},
    'numVotes': {'mean': data['numVotes'].mean(), 'median': data['numVotes'].median()},
}

# Hiển thị thống kê mô tả
print("Thống kê mô tả tổng quan về dữ liệu:")
print(desc_stats)

print("\nGiá trị cao/thấp nhất:")
print(high_low_values)

print("\nGiá trị phổ biến/hiếm nhất:")
print(common_rare_values)

print("\nGiá trị trung bình và trung vị:")
print(average_median_values)
```

```python
# Biểu đồ phân phối Đánh Giá Trung Bình
plt.figure(figsize=(10, 6))
plt.hist(data['averageRating'], bins=30, edgecolor='black')
plt.title('Phân Phối Đánh Giá Trung Bình')
plt.xlabel('Đánh Giá Trung Bình')
plt.ylabel('Tần Suất')
plt.grid(True)
plt.show()

# Biểu đồ phân phối Thời Lượng Phim
plt.figure(figsize=(10, 6))
plt.hist(data['runtimeMinutes'].dropna(), bins=30, edgecolor='black')
plt.title('Phân Phối Thời Lượng Phim')
plt.xlabel('Thời Lượng Phim (Phút)')
plt.ylabel('Tần Suất')
plt.grid(True)
plt.show()

# Biểu đồ phân phối Số Lượng Phiếu Bầu
plt.figure(figsize=(10, 6))
plt.hist(data['numVotes'], bins=30, edgecolor='black', log=True)  # log scale due to wide range
plt.title('Phân Phối Số Lượng Phiếu Bầu')
plt.xlabel('Số Lượng Phiếu Bầu')
plt.ylabel('Tần Suất (log scale)')
plt.grid(True)
plt.show()
```
